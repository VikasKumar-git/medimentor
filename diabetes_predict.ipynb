```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_diab=pd.read_csv("ml_datasets/diabetes.csv")
```

```python
df_diab
```

```python
len(df_diab)
```

```python
len(df_diab[(df_diab['SkinThickness'] == 0)])
```

```python

avg_skin=[]

curr_age=0
for i in range(0,8):
    avg_skin__ = df_diab[ (df_diab['Age'] > curr_age) & (df_diab['Age'] <= curr_age+10) & (df_diab['SkinThickness'] != 0) ] ['SkinThickness'].mean()
    avg_skin.append(avg_skin__)
    curr_age+=10

print(avg_skin)
curr_age=0
for i in range(0,8):
    df_diab.loc[(df_diab['Age'] >curr_age ) & (df_diab['Age'] <= curr_age+10) & (df_diab['SkinThickness'] == 0), 'SkinThickness'] = avg_skin[i]
    curr_age+=10
```

```python
bins = [0.0, 0.2, 0.5, 0.9,3.0]  
labels = ['NoHistory','LowHistory', 'ModerateHistory', 'HighHistory']
df_diab['DiabetesPedigreeFunction'] = pd.cut(df_diab['DiabetesPedigreeFunction'], bins=bins, labels=labels, include_lowest=True)
```

```python
df_diab.describe()
```

```python
bins = [40, 90, 140, 200,394]  
labels = ['Low_intake', 'Medium_intake', 'High_intake','Very_high_intake']
df_diab['glucose_category'] = pd.cut(df_diab['Glucose'], bins=bins, labels=labels, include_lowest=True)
```

```python
df_diab
```

```python
df_diab=df_diab.drop(['Glucose'],axis="columns")

df_diab
```

```python
df_diab.DiabetesPedigreeFunction=df_diab.DiabetesPedigreeFunction.replace({'NoHistory':0, 'LowHistory':1, 'ModerateHistory':2,'HighHistory':3})
```

```python

df_diab.BloodPressure=df_diab.BloodPressure/df_diab.BloodPressure.max()

df_diab.SkinThickness=df_diab.SkinThickness/df_diab.SkinThickness.max()

df_diab.Insulin=df_diab.Insulin/df_diab.Insulin.max()

df_diab.BMI=df_diab.BMI/df_diab.BMI.max()

df_diab.Age=df_diab.Age/df_diab.Age.max()
```

```python
df_diab
```

```python
df_diab.glucose_category=df_diab.glucose_category.replace({'Low_intake':0,'Medium_intake':1,'High_intake':2,'Very_high_intake':3})
```

```python
df_diab
```

```python
len(df_diab)
```

```python
df_diab=df_diab.dropna()
```

```python
df_diab
```

```python
df_diab=df_diab.drop(['Insulin'],axis="columns")
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(df_diab.drop(["Outcome"],axis="columns"),df_diab.Outcome,test_size=0.2)
```

```python
print(len(x_train), len(x_test))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()

model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
