```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_hyper=pd.read_csv("ml_datasets/Hypertension-risk-model-main.csv")
```

```python
df_hyper
```

```python
df_hyper = df_hyper.rename(columns={'male': 'gender'})
```

```python
df_hyper
```

```python
female_chol_avg=df_hyper[df_hyper['gender'] == 0]['totChol'].mean()
male_chol_avg=df_hyper[df_hyper['gender'] == 1]['totChol'].mean()

df_hyper.loc[df_hyper['gender'] == 0, 'totChol'] = df_hyper.loc[df_hyper['gender'] == 0, 'totChol'].fillna(female_chol_avg)

df_hyper.loc[df_hyper['gender'] == 1, 'totChol'] = df_hyper.loc[df_hyper['gender'] == 1, 'totChol'].fillna(male_chol_avg)
```

```python
print(df_hyper.totChol.min(),df_hyper.totChol.max())
```

```python
bins = [107, 303, 500, 696]  
labels = ['Low', 'Medium', 'High']
df_hyper['chol_category'] = pd.cut(df_hyper['totChol'], bins=bins, labels=labels, include_lowest=True)
```

```python
df_hyper
```

```python
df_hyper.chol_category.value_counts()
```

```python
df_hyper=df_hyper.drop(['totChol'],axis='columns')
```

```python
female_glucose_avg=df_hyper[df_hyper['gender'] == 0]['glucose'].mean()
male_glucose_avg=df_hyper[df_hyper['gender'] == 1]['glucose'].mean()

df_hyper.loc[df_hyper['gender'] == 0, 'glucose'] = df_hyper.loc[df_hyper['gender'] == 0, 'glucose'].fillna(female_glucose_avg)

df_hyper.loc[df_hyper['gender'] == 1, 'glucose'] = df_hyper.loc[df_hyper['gender'] == 1, 'glucose'].fillna(male_glucose_avg)

print(female_glucose_avg,male_glucose_avg)
```
