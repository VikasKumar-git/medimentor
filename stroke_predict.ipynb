```python
print(
    df_stroke.gender.value_counts
)
```

```python
female_bmi_avg=df_stroke[df_stroke['gender'] == 'Female']['bmi'].mean()
male_bmi_avg=df_stroke[df_stroke['gender']== 'Male']['bmi'].mean()

print(female_bmi_avg,male_bmi_avg)
```

```python
df_stroke.loc[df_stroke['gender'] == 'Female', 'bmi'] = df_stroke.loc[df_stroke['gender'] == 'Female', 'bmi'].fillna(female_bmi_avg)

df_stroke.loc[df_stroke['gender'] == 'Male', 'bmi'] = df_stroke.loc[df_stroke['gender'] == 'Male', 'bmi'].fillna(male_bmi_avg)
```

```python
df_stroke
```

```python
df_stroke.gender=df_stroke.gender.replace({
    'Male':1,
    'Female':0
})
df_stroke.ever_married=df_stroke.ever_married.replace({
    'Yes':1,
    'No':0
})

df_stroke
```

```python
df_stroke.smoking_status.unique
```

```python
df_stroke = df_stroke[df_stroke['smoking_status'] != 'Unknown']
```

```python
print(df_stroke.smoking_status.value_counts())
```

```python
df_stroke.smoking_status=df_stroke.smoking_status.replace({
    'never smoked': 0,
    'formerly smoked':1,
    'smokes': 2
})
```

```python
df_stroke
```

```python
df_stroke.Residence_type.value_counts()
```

```python
df_stroke.Residence_type.value_counts()
```

```python
df_stroke.Residence_type=df_stroke.Residence_type.replace({'Urban':1, 'Rural':0})
```

```python
df_stroke.work_type.value_counts()
```

```python
df_stroke.age=df_stroke.age.round()
```

```python
df_stroke = df_stroke[df_stroke['age'] != 0]
```

```python
df_stroke
```

```python
df_stroke.work_type.value_counts()
```

```python
df_stroke.work_type=df_stroke.work_type.replace({
    'children': 0,
    'Never_worked':1,
    'Self-employed':2,
    'Govt_job':3,
    'Private':4
})

df_stroke = df_stroke[df_stroke['gender'] != 'Other']
df_stroke=df_stroke.drop(['id'],axis='columns')
df_stroke=df_stroke.drop(['avg_glucose_level'],axis='columns')
```
