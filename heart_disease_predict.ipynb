```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_heart=pd.read_csv("ml_datasets/Heart_disease_cleveland_new.csv")

df_heart
```

```python
df_heart=df_heart.drop(['cp'],axis='columns')
```

```python
df_heart.describe()
```

```python
bins = [107, 303, 500, 696]  
labels = ['Low', 'Medium', 'High']
df_heart['chol_category']=pd.cut(df_heart.chol,bins=bins,labels=labels,include_lowest=True)
```

```python
df_heart
```

```python
df_heart=df_heart.drop(['chol','restecg','oldpeak','slope','ca','thal'],axis='columns')
df_heart.chol_category=df_heart.chol_category.replace({'Low':0,'Medium':1,"High":2})
```

```python
df_heart
```

```python
count=20
avg_heart_rate=[]
for i in range(2,8):
    avg_heart_rate.append(df_heart[(df_heart['age']>count) & (df_heart['age']<count+10)].thalach.mean())
    count+=10

print(avg_heart_rate)

print(type(avg_heart_rate))
```

```python
count=20
for i in range(0,6):
    df_heart.loc[(df_heart['age']>count) & (df_heart['age']<count+10), 'thalach']=avg_heart_rate[i]
    count+=10
```

```python
df_heart
```

```python
df_heart.isna().sum()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_heart.drop(['target'],axis='columns'),df_heart.target,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
